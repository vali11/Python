```python
import numpy as np

data = np.loadtxt('populatii.txt')
an, iepure, linx, morcov = data.T
print(data)
```

```python
print("Media aritmetica pentru iepuri: ", np.mean(data[:,1]))
print("Media artimetica pentru lincsi: ", np.mean(data[:,2]))
print("Media aritmetica pentru morcovi: ",np.mean(data[:,3]))
```

```python
print("Variatia standard pentru iepuri: ",np.std(data[:,1]))
print("Variatia standard pentru lincsi: ",np.std(data[:,2]))
print("Variatia standard pentru morcovi: ",np.std(data[:,3]))
```

```python
sumaIepure = np.sum(data[:,1])
sumaLinx = np.sum(data[:,2])
sumaMorcov = np.sum(data[:,3])
print(sumaIepure, sumaLinx, sumaMorcov)
```

```python
import numpy as np
print("Cea mai mare populatie este de: ",max(np.sum(data[:,1]),np.sum(data[:,2]),np.sum(data[:,3])))
```

```python
print("Anii cu orice populatie mai mare de 50000: ")
print(an[np.any(data[:, 1:]>50000,axis=1)])
```

```python
print("Primii 2 ani pentru fiecare specie în care au avut cele mai mici populații: ")
print(an[np.argsort(data[:, 1:],axis=0)[:2]])
```

```python
import matplotlib.pyplot as plt
iepuri_gradient = np.gradient(iepure)
print("Corelatia dintre iepuri si lincsi: ", np.corrcoef(iepuri_gradient,linx)[0,1])
plt.axes([0.1, 0.1, 0.6, 0.8])
plt.plot(an, iepuri_gradient, label='Gradient iepuri')
plt.plot(an, linx, label='Linx')
plt.legend(('Gradient iepuri', 'Linx'), loc=(1.05, 0.5))
plt.show()
```

```python
plt.axes([0.2, 0.1, 0.5, 0.8])
plt.plot(an, iepure, an, linx, an, morcov)
plt.legend(('Iepure', 'Linx', 'Morcov'), loc=(1.05, 0.5))
plt.show()
```
